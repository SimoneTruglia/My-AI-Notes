si parla della non centratura sullo zero sull'asse delle Y della sigmoide e di como questo sia un problema sull'ottimizzazione dei parametri.

**Enzo dice**


> quando viene utilizzata durante gli aggiornamenti per aggiornare i parametri, la sua derivata prima potrà avere soltanto un segno e quindi i parametri verranno aggiornati soltanto in maniera positiva o negativa. 
>
> Quindi se dovessimo effettuare una serie di aggiornamenti sui nostri pesi, di volta in volta su segni diversi in base agli esempi che vediamo in input, dovremmo in realtà effettuare più aggiornamenti per fare la  singola correzione


Cerchiamo di capire in questo notebook perché la non centratura sullo zero sull'asse delle Y della Sigmoide vada a creare un problema durante la fase di aggiornamento dei pesi

**I singoli componenti $\theta_j$ sono dati da:**

$
\theta_0$ = $\theta_0 - \frac{\alpha}{m} \cdot \left[ \left(h_{\theta}(X^{1}) - Y^{1}\right) \cdot X^{1}{0} + \cdots + \left(h{\theta}(X^{m}) - Y^{m}\right) \cdot X^{m}_{0} \right]
$

$\left[\cdots\right]$

$
\theta_n$ = $\theta_n - \frac{\alpha}{m} \cdot \left[ \left(h_{\theta}(X^{1}) - Y^{1}\right) \cdot X^{1}{n} + \cdots + \left(h{\theta}(X^{m}) - Y^{m}\right) \cdot X^{m}_{n} \right]
$

**Da cui abbiamo che l'equazione che aggiorna i singoli parametri $\theta_j$ è**

$\theta_j$ = $\theta_j - \frac{\alpha}{m} \cdot   \sum\limits_{i=1}^{m} [ \left(h_{\theta}(X^{i}) - Y^{i}\right) \cdot X^{i}_{j} ] $

**I componenti principali della sommatoria sono**

 $\left(h_\theta(X^i) - Y^i \right)$ e $X^i_j$

Analizzando $(h_\theta(X^i) - Y^i)$ vediamo che questo può assumere diversi valori (sulla base della nostra $Y_{ipotesi}$ e della nostra $Y_{target}$) **ma il suo valore sarà sempre uguale, per ognuno dei calcoli sui $j$ parametri** da **0** a **n**



Quindi $(h_\theta(X^i) - Y^i)$ **può essere considerato un valore costante nelle varie equazioni** sui diversi parametri $Θ_j$ e avrà sempre lo stesso sengo (e valore) su ognuno dei parametri $Θ_j$

Gli unici valori dunque in grado di influenzare il segno dei singoli $Θ_j$ sono gli **n** input $X^{m}_{0} ... X^{m}_{n}$



Per concludere quindi, durante l'aggiorament dei parametri $\theta_j$ **gli unici valori che possono influenzare il segno** dei singolo $\theta_j$ sono i vari $X^i_j$ 

Quindi il problema della non centratura sullo zero non è presente sempre in ogni caso ma solo in alcuni casi particolari, quando tutti i valori di X hanno lo stesso segno (tutti positivi o tutti negativi).

Questo porta ad una situazione in cui in ogni singola epoca tutti gli elementi $Θ_j$ vengonoa aggiornati dello stesso segno e, di conseguenza, l'errore si muove a zig-zag sulla funzione di costo
